```r
#Load data
data <- read.table("https://raw.githubusercontent.com/pine-bio-support/Predictive_model_Classification_Regression/main/CellLines_52samples_ExprData_tpose2.txt", sep=",", header=TRUE, check.names=F)

#Check dimensions
dim(data)

#Split training and test data into 80:20 ratio 
dt = sort(sample(nrow(data), nrow(data)*.8))

#Get training dataset
training_set<-data[dt,]

#Get test dataset
test_set<-data[-dt,]


train.x = training_set[2:6917]
train.y = training_set$class
test.x = test_set[2:6917]
test.y = test_set$class
```

```r
install.packages("keras")
install.packages("tensorflow")
install.packages("ggplot2")
install.packages("lattice")
install.packages("caret")
install.packages("dplyr")
#install.packages("mxnet")
install.packages("MASS")


library(keras)
library(tensorflow)
library(ggplot2)
library(lattice)
library(caret)
library(dplyr)
#library(mxnet)
library(MASS)
```

```r
model <- keras_model_sequential() 

  # Add layers to the model
  model %>% 
  layer_dense(units = 8, activation = 'relu', input_shape = c(4)) %>% 
  layer_dense(units = 3, activation = 'softmax')
  # Print a summary of a model
  summary(model)
  
  # Get model configuration
  get_config(model)
  
  # Get layer configuration
  get_layer(model, index = 1)
  
  # List the model's layers
  model$layers
  
  # List the input tensors
  model$inputs
  
  # List the output tensors
  model$outputs
```

```r
# Compile the model
  model %>% compile( loss = 'categorical_crossentropy',optimizer = 'adam', metrics = 'accuracy' )


  # Fit the model 
  model %>% fit(train.x, train.y, epochs = 200,  batch_size = 5,  validation_split = 0.2 )
  
  
  # Store the fitting history in `history` 
  history <- model %>% fit( train.x, train.y, epochs = 200, batch_size = 5,  validation_split = 0.2)
  
  # Plot the history
  plot(history)

  model
  
  
  

  
```

```r
# Plot the model loss of the training data
  plot(history$metrics$loss, main="Model Loss", xlab = "epoch", ylab="loss", col="blue", type="l")
  
  # Plot the model loss of the test data
  lines(history$metrics$val_loss, col="green")
  
  # Add legend
  legend("topright", c("train","test"), col=c("blue", "green"), lty=c(1,1))
  
  
  # Plot the accuracy of the training data 
  plot(history$metrics$acc, main="Model Accuracy", xlab = "epoch", ylab="accuracy", col="blue", type="l")
  
```

```r
  # Plot the accuracy of the validation data
  lines(history$metrics$val_acc, col="green")
  
  # Add Legend
  legend("bottomright", c("train","test"), col=c("blue", "green"), lty=c(1,1))
  
  
  # Predict the classes for the test data
  classes <- model %>% predict_classes(test.x, batch_size = 128)
  
  # Confusion matrix
  table(train.y, classes)
  
  
  # Evaluate on test data and labels
  score <- model %>% evaluate(test.x, test.y, batch_size = 128)
  
  # Print the score
  print(score)
```
